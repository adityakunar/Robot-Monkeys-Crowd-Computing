```python
pair_wise = pd.read_csv("Pairwise_HIT.csv")
```

```python
new_recs=[]
new_labelsss=[]
for i in range(0,100,5):
    onerow = pair_wise[i:i+5]
    oneC = list(onerow["Record1"].values)
    twoC = list(onerow["Record2"].values)
    oneC.extend(twoC)
    label = list(onerow["label"].values)
    label_pairs=[]
    j=0
    for i in label:
        if i==1:
            label=(j,j+5)
            label_pairs.append(label)
        j+=1 
    second = oneC[5:]
    oneCpart2 = []
    for i in enumerate(second):
        oneCpart2.append(i)
    oneCpart2shuffled=shuffle(oneCpart2)
    changelab={}
    for i in enumerate(oneCpart2shuffled):
        changelab[5+i[1][0]]=5+i[0]
    new_labels=[]
    for i in label_pairs:
        new_labels.append((i[0],changelab[i[1]]))
    oneCpartsec = [x[1] for x in oneCpart2shuffled]
    newrec = oneC[:5]
    newrec.extend(oneCpartsec)
    new_recs.append(newrec)
    new_labelsss.append(new_labels)
```
