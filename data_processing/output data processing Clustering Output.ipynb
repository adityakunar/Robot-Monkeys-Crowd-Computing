```python
worker_labels=[]
for i in range(len(outty)):
    clusters=[]
    for j in list(pd.Series(list(outty[i][0].values())).value_counts().index):
        mini_clusters=[]
        for k in list(outty[i][0].keys()):
            if outty[i][0][k]==j:
                mini_clusters.append(int(k))
        if len(mini_clusters)>1:        
            clusters.append(tuple(mini_clusters))
    worker_labels.append(clusters)
```

```python
len(worker_labels)
```

```python
import numpy as np
true_labelss = [[e] * 5 for e in true_labels]
true_labelss=np.array(true_labelss).ravel()
```

```python
worker_labels
```

```python
TP=[]
FP=[]
FN=[]
TN=[]


for i,j in zip(worker_labels,true_labelss):
    
    tp=len(set(i).intersection(j))
    fp=len(set(i)-set(j))
    fn=len(set(j)-set(i))
    tn=256-(tp+fp+fn)
    TP.append(tp)
    FP.append(fp)
    FN.append(fn)
    TN.append(tn)
    
    
```

```python
(np.array(TP).mean()),(np.array(FP).mean()),(np.array(TN).mean()),(np.array(FN).mean())
```

```python
prec= (np.array(TP).mean())/((np.array(TP).mean())+(np.array(FP).mean()))
prec
```

```python
recall=(np.array(TP).mean())/((np.array(TP).mean())+(np.array(FN).mean()))
recall
```

```python
F1_score = 2* (prec*recall)/(prec+recall)
F1_score
```
