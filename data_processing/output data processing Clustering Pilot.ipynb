```python
worker_labels=[]
for i in range(len(outty)):
    clusters=[]
    for j in list(pd.Series(list(outty[i][0].values())).value_counts().index):
        mini_clusters=[]
        for k in list(outty[i][0].keys()):
            if outty[i][0][k]==j:
                mini_clusters.append(int(k))
        if len(mini_clusters)>1:        
            clusters.append(tuple(mini_clusters))
    worker_labels.append(clusters)
```

```python
top5=[]
for i in range(0,len(worker_labels),5):
    top5.append(pd.Series(worker_labels[i:i+5]).value_counts().index[0])
```

```python
top3=[]
for i in range(0,len(worker_labels),5):
    top3.append(pd.Series(worker_labels[i:i+3]).value_counts().index[0])
```
